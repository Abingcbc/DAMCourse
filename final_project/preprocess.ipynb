```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
lines_list = []
points = []
```

```python
with open("/Users/cbc/Project/Python/DAMCourse/final_project/data_package/training_data/0.txt") as file:
    for line in file:
        temp = line.split(" ")
        if len(temp) == 2:
            temp = [eval(temp[0]),eval(temp[1])]
            points.append(temp)
        else:
            lines_list.append(points)
            points = []
    lines_list.append(points)
```
